```python
x = gs.planar(zdim=32, dim=784)
```

```python
plt.scatter(x[:, 0], x[:, 1], s=0.1)
```

```python
rlt = gs.rlts(x, n=100, L_0=64, i_max=10)
```

```python
mrlt = np.mean(rlt, axis=0)
```
